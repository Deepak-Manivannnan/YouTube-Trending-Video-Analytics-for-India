### 4.7 Create Language Column

```python
language_keywords = {
    "English" : ["english"],
    "Hindi": ["hindi"],
    "Tamil": ["tamil"],
    "Telugu": ["telugu"],
    "Malayalam": ["malayalam"],
    "Kannada": ["kannada"],
    "Punjabi": ["punjabi"],
    "Bengali": ["bengali", "bangla"],
    "Marathi": ["marathi"],
    "Gujarati": ["gujarati"],
    "Odia": ["odia", "oriya"]
}

def detect_language(row):
    title = str(row["video_title"]).lower()
    tags = str(row["tags"]).lower()

    for language, keywords in language_keywords.items():
        if any(keyword in title for keyword in keywords):
            return language
        elif any(keyword in tags for keyword in keywords):
            return language

    return "Unknown"

df["language"] = df.apply(detect_language, axis=1)
```

```python
df["language"].value_counts(normalize=True)
```

- 57.7% of videos are classified as Unknown, meaning more than half the dataset could not be mapped to a specific language. This is because language detection relies on keyword matching in video titles and tags — videos without explicit language keywords default to Unknown. 
- Language-based analysis should be interpreted with caution.

### 4.11 Collapse to One Row Per Video

```python
new_df = df.groupby("video_id").agg(
    video_title = ("video_title", "last"),
    published_at = ("published_at", "first"),
    channel_title = ("channel_title", "first"),
    language = ("language", "first"),
    category_id = ("category_id", "first"),
    first_trending_date = ("trending_date", "min"),
    last_trending_date = ("trending_date", "max"),
    view_count = ("view_count", "max"),
    likes = ("likes", "max"),
    dislikes = ("dislikes", "max"),
    comment_count = ("comment_count", "max"),
    comments_disabled = ("comments_disabled", "first"),
    ratings_disabled = ("ratings_disabled", "first")
).reset_index()
```

```python
new_df["days_to_trend"] = (new_df["first_trending_date"].dt.date - new_df["published_at"].dt.date).apply(lambda x : x.days) 
```

```python
new_df["days_trending"] = (new_df["last_trending_date"] - new_df["first_trending_date"]).dt.days + 1
```

```python
new_df.head()
```


- The raw dataset contains multiple rows per video — one for each day it appeared on the trending page. The engagement metrics (view_count, likes, dislikes, comment_count) are cumulative, meaning each row includes all previous days' totals. Keeping multiple rows per video causes any aggregation to produce inflated, misleading results.

- To fix this, each video is consolidated into a single row using video_id as the unique identifier, retaining the final cumulative metric values. The grouping also captures:
  - `first_trending_date` — the first day the video appeared on trending
  - `last_trending_date` — the last day the video appeared on trending

- Two additional columns are then derived from the above:
  - `days_to_trend` — days between publish date and first trending date
  (0 means the video trended on the same day it was published)
  - `days_trending` — total days the video stayed on the trending page

- After collapsing, the dataset contains one row per unique video, making all downstream analysis accurate and consistent.
